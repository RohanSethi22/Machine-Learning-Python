```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10,10)
y = x**2

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y)

def onclick(event):
    global ix, iy
    ix, iy = event.xdata, event.ydata
    print('x = ', iy)
    ax.plot(ix,iy)

    global coords
    coords = [ix, iy]

    return coords


for i in range(0,1):

    cid = fig.canvas.mpl_connect('button_press_event', onclick)


plt.show()
```
